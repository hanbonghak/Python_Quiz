```python
count = 0
for i in range(1000,10000) :
    if '10' in str(i)  : 
        count += 1
print('\'10\'을 포함하는 숫자의 갯수 =',count)
```

```python
count = 0
for i in range(10,100000) :
    if '080' in str(i) and i%20==0 :
        count += 1
print('20의 배수이며 \'080\'을 포함하는 숫자의 갯수 =',count)
```
